# 9. How do you fill missing numerical values using the mean?

```python
df = df.fillna(df.mean(numeric_only=True)) 
df
```

# 10. How do you convert age from days to years?

```python
df['age'] = df['age'] // 365
df
```

# 11. How do you calculate the correlation matrix?

```python
df.corr()   
```

# 12. How do you check the distribution of the target variable (cardio)?

```python
df['cardio'].value_counts()
```

# 13. How do you count how many people have high cholesterol?

```python
len(df[df['cholesterol'] == 2])    
```

# 14. How do you calculate the percentage distribution of glucose levels?

```python
df['gluc'].value_counts(normalize=True) 
```

# 15. How do you analyze the relationship between smoking and cardiovascular disease?

```python
df.groupby('smoke')['cardio'].mean()   
```

# 16. How do you find the number of duplicated rows?

```python
df.duplicated().sum()
```

# 17. How do you remove duplicated rows (if any)?

```python
df = df.drop_duplicates()
```

# 18. How do you plot the distribution of weight?

```python

import matplotlib.pyplot as plt  
df['weight'].hist()
plt.show()
```

# 19. How do you calculate the average blood pressure?

```python
df['bp_avg'] = (df['ap_hi'] + df['ap_lo']) / 2 
df['bp_avg'].head()
```

# 20. How do you find the feature most correlated with the target (cardio)?

```python
df.corr()['cardio'].sort_values(ascending=False)   
```

# 21. How do you plot the distribution of blood pressure (ap_hi)?

```python
df['ap_hi'].hist()    
plt.xlabel("Systolic BP")
plt.ylabel("Count")
plt.show()
```
